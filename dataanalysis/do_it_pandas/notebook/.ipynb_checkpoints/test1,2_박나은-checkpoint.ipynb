```python
# 앞면이 나올 확률이 0.4인 동전이 있다. 
# 이 동전을 5번 던져서, 앞면이 2회 이하 나올 확률은 얼마인가?

zero = (3/5)**5          #0번 나올 경우
# 뒷면이 5번 나오는 횟수가 0번이므로 1을 곱한다

one = (2/5)*(3/5)**4*5      #1번 나올 경우
# 앞면 한 번, 뒷면 네 번이 나오는 확률에 앞 면이 어디에 나오는지(5!를 4!로 나눈 값)에 따라 총 5가지이므로 5를 곱한다

two = (2/5)*(2/5)*(3/5)**3*10   #2번 나올 경우
# 앞면 두 번, 뒷면 세 번이 나오는 확률에 앞 면이 어디에 나오는지(5!을 3!2!로 나눈 값)에 따라 총 10가지이므로 10을 곱한다

zero + one + two
```

```python
# tips 데이터에서 size가 3 ~ 5인 데이터에 대해, time별 tiprate의 평균은?
# 단, tiprate는 tip/total_bill이다.

import numpy as np
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips = tips[tips['size'].isin(['3', '4', '5'])]
```

```python
tips[ 'tiprate' ] = tips[ 'tip' ] / tips[ 'total_bill' ]
tips
```

```python
tips = tips.groupby(['time']).mean()

tips
```
