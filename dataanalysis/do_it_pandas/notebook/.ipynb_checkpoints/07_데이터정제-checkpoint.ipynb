## 7.1 melt 메소드 사용하기(151쪽)
melt 메소드는 wide 데이터에서 지정한 열을 모두 행으로 전환해준다. 이때 wide 데이터의 저장 순서에 따라 적재가 이뤄진다.

![](https://t1.daumcdn.net/cfile/tistory/25177F4E5863D58A0C)

pd.melt(frame, id_vars=None, value_vars=None, var_name=None, value_name='value', col_level=None):
- id_vars: list-like. 위치를 그대로 유지할 열 이름들을 지정한다.
- value_vars: list-like. 행으로 위치를 변경할 열의 이름을 지정한다.
- var_name: str. 행으로 위치를 변경되는 열 이름들을 저장할 variable 열 이름을 지정한다.
- value_name: str. value_vars의 값인 value를 저장할 열 이름을 지정한다.
- col_level: columns가 MultiIndex인 경우, melt할 level을 지정한다.

```python
tmp = pew_long[~pew_long.income.str.startswith('Do')]
incval = [5, 15, 25, 35, 45, 62.5, 87.5, 125, 150]
incdict = {}
for val, inc in zip(incval, tmp.income.unique()):
    incdict[inc] = val
    
tmp['inc'] = tmp.income.map(lambda x: incdict[x])
tmp['incsum'] = tmp['inc'] * tmp['count']
tmp
```

```python
tmpsum = tmp.groupby('religion')[['count', 'incsum']].sum()
tmpsum
```

```python
tmpsum['incmean'] = tmpsum['incsum'] / tmpsum['count']
tmpsum
```

### 7.2.3 concat 메소드를 응용하여 데이터프레임에 열 추가하기(159쪽)
concat 메소드를 이용하면 split 메소드로 분리한 데이터를 바로 데이터프레임에 추가할 수 있다.

expand=True 옵션을 사용하면, Series가 아닌 DataFrame을 리턴한다.

```python
variable_split = ebola_long.variable.str.split('_', expand=True) 
variable_split.columns = ['status', 'country'] 
ebola_parsed = pd.concat([ebola_long.iloc[:, :-2], variable_split], axis=1)

ebola_parsed[100:160]
```

```python
ebola_parsed.drop('variable', 1).fillna(0)
```

위에서 한 것과 같이 fillna는 적절하지 않다.

[Quiz] ebola_parsed에서:
1. variable 변수를 제거하고,
2. status와 country로 그룹하여
3. 날짜를 오름차순으로 정렬하고
4. 결측치를 행 방향으로 제거한 temp1
5. 결측치를 ffill로 대체한 temp2 데이터프레임을 생성한다.

```python
ebola_parsed.drop('variable', 1, inplace=True)
```

```python
manyExe = lambda x: x.sort_values('Day').dropna()
```

groupby 객체는 DF 객체가 아니기 때문에, dropna와 같은 메소드가 없다.

```python
%%timeit
temp1 = ebola_parsed.groupby(['status', 'country']).apply(manyExe)
temp1
```

apply를 않쓰려면 for문으로 처리한다.

```python
%%timeit
tmp = []
for _, group in ebola_parsed.groupby(['status', 'country']):
    tmp.append(group.sort_values('Day').fillna(method='pad'))

temp2 = pd.concat(tmp)
```

## 7.3 여러 열을 하나로 정리하기
비슷한 정보를 갖는 여러 열을 하나로 통합하는 방법을 알아보자.
### 7.3.1 기상 데이터의 여러 열을 하나로 정리하기(160쪽)
다음은 기상 데이터를 불러와 출력한 것입니다. 날짜 열(d1, ..., d31)에는 각 월별 최고, 최저 온도 데이터가 저장되어 있습니다. 지금은 날짜 열이 옆으로 길게 늘어져 있어 보기 불편합니다.

```python
weather = pd.read_csv('../data/weather.csv') 
print(weather.iloc[:5, :11])
```

#### 1) melt 이용하기
다음은 melt 메서드로 일별 온도 측정값(d1,d2,...)을 피벗한 것입니다. 

그러면 day 열에 날짜 열이 정리되고 날짜 열의 데이터는 temp 열에 정리됩니다. 하지만 아직 최고, 최저 온도가 한눈에 잘 들어오지 않습니다.

```python
weather_melt = pd.melt(weather, id_vars=['id', 'year', 'month', 'element'], 
                       var_name='day', value_name='temp') 
weather_melt.head()
```

```python
weather_melt[:30]
```

#### 2) pivot_table 이용하기
이제 melt한 데이터프레임에 pivot_table 메서드를 적용할 차례입니다. pivot_table 메서드는 행과 열의 위치 를 다시 바꿔 정리해 줍니다.
![](https://t1.daumcdn.net/cfile/tistory/255C7950585E784E01)
- index 인자에는 위치를 그대로 유지할 열 이름을 지정하고, 
 - index로 지정된 열들은 순서대로 계층적 index를 구성한다.
- columns 인자에는 피봇 열들을 구성할 열 이름을 지정하고, 
 - columns에는 범주형 열을 지정하며, 그 범주들은 새로운 열이름이 된다.
- values 인자에는 새로운 열의 데이터가 될 열 이름을 지정하면 됩니다. 

다음은 pivot_table 메서드로 행과 열의 위치를 다시 바꾼 것 입니다. 위의 결과와 비교해보세요.

```python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'], 
    columns='element', 
    values='temp'
)

weather_tidy
```

```python
weather_tidy.shape
```

- tmax와 tmin이 모두 결측인 row는 자동으로 제거된다.
- 다른 변수들은 모두 색인으로 처리되므로, 모든 열(tmax, tmin)이 결측이면 제거하는 것이 합리적이다.

```python
weather_tidy_flat = weather_tidy.reset_index() 
weather_tidy_flat.head()
```

```python
weather_tidy_flat.day = weather_tidy_flat.day.map(lambda x: int(x[1:]))
weather_tidy_flat.head()
```

```python
weather_tidy_flat.dtypes
```

이제 데이터를 정렬하고, 인덱스를 리셋하면:

```python
weather_tidy_flat = weather_tidy_flat.\
    sort_values(['year', 'month', 'day']).reset_index(drop=True)
weather_tidy_flat
```

날짜와 관련된 정보를 datetime 포맷으로 변경하자.

```python
weather_tidy_flat['date'] = pd.to_datetime(weather_tidy_flat[
    ['year', 'month', 'day']])
weather_tidy_flat = weather_tidy_flat.drop(
    ['year', 'month', 'day'], 1)

weather_tidy_flat
```

1. 심야할증이 적용되는 00:00:00에서 05:00:00 데이터만을 선택하고
2. locationID가 결측된 데이터를 우선 제외하여 결합하는 코드를 작성하라.

```python
taxi.dtypes
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
```

```python
list_taxi_df = []

for csv_filename in nyc_taxi_data:
    # print(csv_filename)
    df = pd.read_csv(csv_filename, parse_dates=['Pickup_date'], 
                     date_parser=dateparse)
    hours = df.Pickup_date.map(lambda x: x.hour)
    condts = (hours < 5) & (df.locationID.notna())
    df = df[condts]
    list_taxi_df.append(df) 

print(len(list_taxi_df))
```

```python
taxi_loop_concat = pd.concat(list_taxi_df, ignore_index=True) 
print(taxi_loop_concat.shape)
```

```python
ebola_l = ebola.melt(id_vars=['Date', 'Day'])
ebola_l.head()
```

```python
tmp = ebola_l['variable'].str.split('_', expand=True)
tmp.columns = ['status', 'Country']
tmp = pd.concat([ebola_l.drop('variable', 1), tmp], axis=1)
tmp.dropna(inplace=True)
tmp.head()
```

```python
ebola_wide = tmp.pivot_table(
    index=['Date', 'Day', 'Country'],
    columns='status',
    values='value'
).reset_index()

ebola_wide.columns.name = ''
ebola_wide.head()
```

```python
cols = list(ebola_wide.columns)[-2:]
cols
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, axes = plt.subplots(4,2, sharex=True,sharey=True,
                        figsize=(10,20))

meanDeathRate = {}
for i, (nation, group) in enumerate(ebola_wide.groupby('Country')):
    group.sort_values('Day', inplace=True)
    group.fillna(method='pad', inplace=True)
    group['deathRate'] = group.Deaths / group.Cases
    group.set_index('Date')[cols].plot(ax=axes[i//2][i%2])
    axes[i//2][i%2].set_title(nation)
    
    meanDeathRate[nation] = group['deathRate'].mean()
```

```python
barfig, ax1 = plt.subplots(figsize=(12,8))

pd.Series(meanDeathRate).plot(kind='bar', ax=ax1)
```
