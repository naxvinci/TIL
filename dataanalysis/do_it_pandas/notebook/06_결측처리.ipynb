## 6.1 결측값이란
결측값은 NaN, NAN, nan 등으로 표기되며, 교재에서는 NaN으로 표기한다.
### 6.1.1 결측값 확인하기(135쪽)
결측값 확인을 위해 numpy에서 결측값 객체를 가져온다.

```python
from numpy import NaN, NAN, nan
```

결측값은 말 그대로 결측값이며, 그 어떤 다른 object와 같지 않다.

```python
print(NaN == True)
```

```python
print(NaN == False)
```

```python
print(NaN == 0)
```

```python
print(NaN == '')
```

자기자신과 비교해도 True가 아닌 False를 리턴한다.

```python
print(NaN == NaN)
```

```python
print(NaN == nan)
```

```python
print(NaN == NAN)
```

```python
print(nan == NAN)
```

결측값을 확인하는 함수 또는 메소드: isnull, isna, notnull, notna

```python
import pandas as pd 

print(pd.isnull(NaN))
```

```python
print(pd.isna(nan))
```

```python
print(pd.isnull(NAN))
```

```python
print(pd.notnull(NaN))
```

```python
print(pd.notna(42))
```

```python
print(pd.notnull('missing'))
```

### 6.2.1 결측값의 개수 구하기(142쪽)

```python
ebola = pd.read_csv('../data/country_timeseries.csv')
ebola
```

```python
print(ebola.shape)
print(ebola.count())
```

row 수에서 count의 수를 빼면 결측값의 수를 확인할 수 있다.

```python
num_rows = ebola.shape[0]
num_missing = num_rows - ebola.count() 
print(num_missing)
```

데이터 전체에서의 결측치의 개 수

```python
import numpy as np 

print(np.count_nonzero(ebola.isnull()))
```

```python
ebola.isna().sum().sum()
```

### 6.2.2 결측값 처리하기 ― 대체, 삭제(144쪽)
결측값은 삭제하거나(drop)나 대체(impute)할 수 있다.

#### 1) 결측값 대체하기

```python
ebola.Cases_Guinea.mean()
```

```python
ebola.fillna(0)
```

```python
ebola.fillna(0).Cases_Guinea.mean()
```

```python
ebola.fillna(0).iloc[0:10, 0:5]
```

```python
print(ebola.fillna(method='ffill').iloc[0:10, 0:5])
```

```python
print(ebola.fillna(method='bfill').iloc[-10:, 0:5])
```

```python
ebola.interpolate().iloc[0:10, 0:5]
```
