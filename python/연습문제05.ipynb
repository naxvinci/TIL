```python
word_dict = dict()
for word in words:
    if word in word_dict:
        word_dict[word] = word_dict[word] + 1
    else:
        word_dict[word] = 1
word_dict
```

```python
word_dict = dict()
for word in words:
    word_dict[word] = words.count(word)
word_dict
```

```python
import operator

word_list = sorted(word_dict.items(), 
       key=operator.itemgetter(1), reverse = True) 
#튜플리스트 중 두번째 인덱스를 오름차순으로 정리하겠다는 의미// reverse는 그 반대
```

```python
class EachSum:
    def __init__(self):
        self.total = 0
    def compute(self, num):
        while True:
            self.total += num % 10
            num //= 10
            if num == 0: break
        return self.total    

 # 코드 작성
each = EachSum()
total = each.compute(12345)
print('합계', total)
```

```python
class Car:
    
    
     # 생성자 작성
    def __init__(self):
        self.options = []
        self.options = list()
        
     # add_option 메소드 작성    
    def add_option(self, option):
        self.options.append(option)
    
     # show_option 메소드 작성
    def show_option(self):
        return self.options





car1 = Car()
car2 = Car()
car1.add_option('전동 트렁크')
car1.add_option('통풍 시트')
car2.add_option('뒷자리 에어백')
car2.add_option('하이패스')
print(car1.show_option())
print(car2.show_option())
```

```python
class Letter:
    letter = ''
    
    # 변수 선언 및 초기화
    # write 메소드(함수) 작성
    
    def write(self, text): 
        self.letter += text
        
    # show 메소드(함수) 작성
    def show(self):
        return self.letter
    
#생성된 인스턴스의 주소 저장
letter1 = Letter()

#write 호출시 주소가 self로 전달
letter1.write('가나다라')

letter1.write('카타파하')
letter2 = Letter()
letter2.write('ABCD')
letter2.write('WXYZ')
print(letter1.show())
print(letter2.show())
```
