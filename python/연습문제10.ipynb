### 대통령 선거 개표현황 크롤링(nth-child를 활용한 그래프 크롤링)

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver as wd
import time

driver = wd.Chrome(executable_path='chromedriver.exe')
driver.get(' http://info.nec.go.kr/main/showDocument.xhtml?electionId=0000000000&topMenuId=VC&secondMenuId=VCCP09')

driver.find_element_by_id('electionType1').click()

driver.find_element_by_name('electionName').send_keys('제19대')
time.sleep(1)
driver.find_element_by_name('electionCode').send_keys('대통령 선거')
time.sleep(1)
text = driver.find_element_by_css_selector('#cityCode > option:nth-child(2)').text #두번째 옵션값(전체항목)
driver.find_element_by_name('cityCode').send_keys(text)
time.sleep(1)
driver.find_element_by_id('searchBtn').click()
driver.implicitly_wait(2)
```

```python
#그래프에서 자료 추출하기
tr_list = [2, 3, 6, 7, 9, 18]
td_list = [4, 5, 6]

for rr in tr_list:
    tr = driver.find_element_by_css_selector('#table01 > tbody > tr:nth-child(%d)' % rr) # nth-child는 1부터 시작. 티바디의 r번째 tr이란의미
    for dd in td_list: 
        result = tr.find_element_by_css_selector('td:nth-child(%d)' % dd).text
        print(result.split("(")[0].replace(',','')) #출력 포맷 예쁘게하기
```

### 셀레니움을 이용한 인터파크 투어 해외여행 크롤링 (종합) 어려움

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver as wd

driver = wd.Chrome(executable_path='chromedriver.exe')
driver.get('http://tour.interpark.com')

driver.find_element_by_id('SearchGNBText').send_keys('달랏') # select 요소를 선택할 때 이런 코드를 쓰면 된다(드롭바)
driver.find_element_by_class_name('search-btn').click()

driver.implicitly_wait(5)

driver.find_element_by_css_selector('.panelZone > .oTravelBox .moreBtn').click() # 더보기버튼 클릭. 셀레니움 execute를 쓰면 좀 더 수월

import time
time.sleep(2)
```

```python
#요 햠수 새로 수정
def get_html(url):
    html = ''
    res = requests.get(url)
    if res.status_code == 200:
        res.encoding = None #인코딩을 None이라는 데이터로 바꿔준다
        html = res.text
    return html
```
