### dictionary 형태로 변환하여 출력하기

```python
import pymysql
conn = pymysql.connect(
 host='localhost', user='root', password='1234',
 db='mydb', charset='utf8')
# 코드 작성
cursor = conn.cursor(pymysql.cursors.DictCursor) # dictionary 형태의 cursor로 바꿔주기
sql = '''SELECT year, area1, count(*)
            FROM accident
            GROUP BY area1, year'''
cursor.execute(sql) #sql실행
result = cursor.fetchall()
for row in result:
    print(row['area1'], row['year'], row['count(*)'])# 순서 상관없이 원하는 항목 끄집어내기

conn.commit() #저장
cursor.close()
conn.close() # 요런 형태를 더 많이 활용한다 dict형태..
```

### GROUP BY 활용하기

```python
import pymysql
conn = pymysql.connect(
 host='localhost', user='root', password='1234',
 db='mydb', charset='utf8')
# 코드 작성
cursor = conn.cursor() # Cursor 객체 생성
sql = '''SELECT year, area1, count(*)
            FROM accident
            GROUP BY area1, year'''
cursor.execute(sql) #sql실행
result = cursor.fetchall()
for row in result:
    print(row[0], row[1], row[2])

conn.commit() #저장
cursor.close()
conn.close() # Try 내의 지역변수
```

```python
import pymysql

conn = None

try: 
    conn = pymysql.connect(
        host = 'localhost', user = 'root', password = '1234',
        db = 'python', charset = 'utf8') #데이터베이스 접속

    with conn.cursor() as cursor: # Cursor 객체 생성
        sql = '''CREATE TABLE mysql (
                    id INTEGER PRIMARY KEY AUTO_INCREMENT
                    , title VARCHAR(100), content VARCHAR(100))'''
        cursor.execute(sql) #sql실행
        conn.commit() #저장
    
except pymysql.InternalError as e:
    print(e)
    
finally: 
    if conn:
        conn.close() # Try 내의 지역변수
```
